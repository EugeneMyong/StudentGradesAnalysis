# Abstract

In this project, our group aimed to identify what factors most influence high school students’ academic performance. We focused on both personal and mental health factors-such as family relationships, free time and health status-and academic factors like weekly study time, school absences and past class failures. To assess these influences, we analyzed students’ initial and final grades so that we could see if there is any change in performance-and whether that change seems to have a statistically significant cause. Our team hypothesized that personal and mental health factors would have more of an impact on students' grades than academic factors.

To test this hypothesis, we analyzed a dataset of students in two Portuguese high schools. We used linear regression to determine the statistical significance of each factor in relation to both students' initial and final grades, so that we could see if there was improvement in between the two grades. Our research proved our hypothesis to be incorrect. The only variable that showed statistical significance was the number of classes failed. These results suggest that academic factors, particularly past performance, have a greater influence on students' final grades. However, not all academic factors were significant. For example, the amount of time students spent studying did not influence their performance. 

This study provides important insights into the factors that contribute to academic performance, some of which have a greater effect than others. We encourage further research into the complex relationships between personal and academic influences on students' grades. There are likely additional variables, such as sleep patterns or holding part-time jobs that also play a significant role in academic outcomes and deserve further research.


# Research Question

Do personal and mental health factors (family relationships, free time, and health status) or academic factors (weekly study time, school absences, and past class failures) have a greater influence on high school students' final grades (G3)? Additionally, we will compare the influence of these factors on both students' final grades (G3) and their initial grades (G1) to assess any changes over time.

To everyone's surprise, however, many students performed better than when they were in person, which may have been due to open-book tests becoming more common as well as professor flexibility with curves and grades overall, as discovered by a previous COGS 108 group.<a name=”cite_ref-6”></a>[<sup>6</sup>](#cite_note-6)

There are so many reasons why some students perform better on exams, and understanding these reasons gives a broader perspective for students to decide how they want to perform well in school by understanding what does influence their performance. # Hypothesis


We hypothesize that academic performance among high school students, as measured by their initial grade (G1) and final grade (G3), is more strongly influenced by personal and mental health factors, such as family relationships, free time, and health status. Academic factors like weekly study time, school absences, and past failed classes are expected to have a lesser impact. By focusing on students' starting performance compared to their final outcomes we are able to get better insights into the factors that contribute to academic growth or decline.

Furthermore, our background research suggests that personal and mental health factors have an impact on academic outcomes. For example, students struggling with poor mental health may have inconsistent school attendance and ineffective study habits, which can amplify their risk of failing classes. By analyzing the relative influence of these variables on students' initial and final grades, we can pinpoint the key drivers of academic success and offer actionable recommendations for improvement.

## Dataset

Dataset Name: Student Performance-Mathematics

Dataset Link: https://archive.ics.uci.edu/dataset/320/student+performance

Number of observations: 395

Number of variables: 33

Student Performance-Mathematics measures the student performance of two Portuguese high schools in a mathematics class. Personal factors are categorized as variables that are unrelated to school and involving personal life. Academic factors are categorized as variables directly involved and related to school. These variables will be analyzed with student performance to assess which factors influence student performance more significantly. All variables are analyzed with each other as integer data types which needs to be standardized across the variables during data cleaning and pre-processing. Average change in grade also needs to be calculated using G1 and G3 grades to measure student grades over the year.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format ='retina'
import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf

from scipy.stats import pearsonr, norm, ttest_ind, skewnorm
```

The dataset is currently not in the ideal format, so we proceed to Data Preparation to make it usable.

## Data Preparation

```python
# Separate column names
columns = df.columns
columns = columns[0].split(";")
columns
```

```python
# Separate data
def reorganize_df(row):
    data = row.iloc[0]
    data = data.split(";")

    for i in range(len(columns)):
        row[columns[i]] = data[i]
    return row
```

```python
df = df.apply(reorganize_df, axis=1)
df = df.iloc[:,1:]
df.head()
```

The data frame is separated into columns and data to make it usable.

## Data Cleaning/Wrangling

```python
# Remove leading and trailing quotes from all cells
df = df.map(lambda x: x.strip('"') if isinstance(x,str) else x)
df.head()
```

```python
# Convert number strings to integers
integer_columns = ['age', 'Medu','Fedu','traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3']
df[integer_columns] = df[integer_columns].apply(lambda x: x.apply(int))
```

Having multiple datatypes isn't a problem, but it would be nice to have integers wherever possible. We replace every 'yes' with 1 and 'no' with 0, but we'll leave the other strings as their base versions because they convey relevant information.

```python
# Convert 'yes' and 'no' to binary values
df = df.map(lambda x: 1 if x == 'yes' else (0 if x == 'no' else x))
```

```python
# Check for missing values
df.isna().any()
```

We then create a new column called Average Grade Change, which tracks the change in grades from G1 (period 1) to G3 (year's end).

```python
# Calculate and add average grade change column
df['Average_Grade_Change'] = df.apply(lambda x: (x['G3'] - x['G1']) / 2, axis=1)
df['Average_Grade_Change']
```

Because we are trying to see whether academic or personal/mental health factors have a bigger impact, we split the dataframe into two smaller dataframes to make it easier to see the impact of different factors.

```python
# Categorize academic and health factors
academic_factors = df[ ["studytime", "failures", "absences", "Average_Grade_Change", "G1", "G3"] ]
health_factors = df[ [ "famrel", "freetime", "health", "Average_Grade_Change", "G1", "G3"] ]

academic_factors
```

We proceed to use Ordinary Linear Regression to see if any of the variables have a significant impact on the final grade.

```python
# Academic Factor: Time Studied
Time_Studied_outcome, Time_Studied_predictors = patsy.dmatrices('Final_Grade ~ Time_Studied', academic_factors)
Time_Studied_model = sm.OLS(Time_Studied_outcome, Time_Studied_predictors)
Time_Studied_results = Time_Studied_model.fit()
print(Time_Studied_results.summary())
```

With a p-value of 0.052, we conclude that Time_Studied is not significantly correlated with Final_Grade.

```python
# Academic Factor: Classes Failed
Classes_Failed_outcome, Classes_Failed_predictors = patsy.dmatrices('Final_Grade ~ Classes_Failed', academic_factors)
Classes_Failed_model = sm.OLS(Classes_Failed_outcome, Classes_Failed_predictors)
Classes_Failed_results = Classes_Failed_model.fit()
print(Classes_Failed_results.summary())
```

With a p-value of 0.00, we conclude that Classes_Failed is significantly correlated with Final_Grade.

```python
# Academic Factor: Absences
Absences_outcome, Absences_predictors = patsy.dmatrices('Final_Grade ~ Absences', academic_factors)
Absences_model = sm.OLS(Absences_outcome, Absences_predictors)
Absences_results = Absences_model.fit()
print(Absences_results.summary())
```

With a p-value of 0.497, we conclude that Absences is not significantly correlated with Final_Grade.

```python
# Personal/Mental Health Factor: Family_Relationship_Rating
Family_Relationship_Rating_outcome, Family_Relationship_Rating_predictors = patsy.dmatrices('Final_Grade ~ Family_Relationship_Rating', health_factors)
Family_Relationship_Rating_model = sm.OLS(Family_Relationship_Rating_outcome, Family_Relationship_Rating_predictors)
Family_Relationship_Rating_results = Family_Relationship_Rating_model.fit()
print(Family_Relationship_Rating_results.summary())
```

With a p-value of 0.309, we conclude that Family_Relationship_Rating is not significantly correlated with Final_Grade.

```python
# Personal/Mental Health Factor:	Free_Time_Rating
Free_Time_Rating_outcome, Free_Time_Rating_predictors = patsy.dmatrices('Final_Grade ~ Free_Time_Rating', health_factors)
Free_Time_Rating_model = sm.OLS(Free_Time_Rating_outcome, Free_Time_Rating_predictors)
Free_Time_Rating_results = Free_Time_Rating_model.fit()
print(Free_Time_Rating_results.summary())
```

With a p-value of 0.823, we conclude that Free_Time_Rating is not significantly with Final_Grade.

```python
# Personal/Mental Health Factor:	Health_Status
Health_Status_outcome, Health_Status_predictors = patsy.dmatrices('Final_Grade ~ Health_Status', health_factors)
Health_Status_model = sm.OLS(Health_Status_outcome, Health_Status_predictors)
Health_Status_results = Health_Status_model.fit()
print(Health_Status_results.summary())
```

With a p-value of 0.224, we conclude that Health_Status is not significantly correlated with Final_Grade.

# Discussion and Conclusion




Our initial goal was to find out whether academic factors (weekly study time, school absences, and past class failures) or personal and mental health factors (family relationships, free time, and health status) have a larger impact on the academic performance of students, as measured by their end of year grades. We hypothesized that personal and mental health factors would have a larger impact on the students performance, as previous work done by others suggests that circumstances involving personal factors generally have more of an impact than any academic factors. Upon plotting the bargraphs of the six variables, however, it becomes evident that despite the general consensus agreeing that factors such as family relationships or study time strongly affect a student's performance, we may have overestimated how impactful these things can be.

Indeed, as discovered through Linear Regression on the six variables chosen and the final grade, the only variable that had a statistically significant impact on the grades of students was the number of classes failed. Weekly study time also had a strong impact on student grades. Thus, our hypothesis was incorrect, it appears that academic factors have more of an impact on students performance. This was not without interesting results, though, as we made many discoveries in our analysis. Not only was it surprising to find out that an academic factor was the most influential factor of a student's grade, but we also discovered that variables such as study time and free time that are commonly agreed to affect a student's grade had no statistically significant impact.

Based on what was seen in the heat maps, the majority of the factors are independent of each other, which is portrayed by the lack of diagonal gradience in the diagrams. The exceptions to this include 'Free Time Rating' vs. 'Time Studied', 'Health Status' vs. 'Time Studied', and 'Free Time Rating' vs. 'Classes Failed', where these pairs of factors are correlated to each other. However, one factor of each pair is categorized as an academic factor, while the other is categorized as a personal and mental factor, so these exceptions are not what we are looking for. Because of this, our hypothesis is not supported.

In our line plots, we found that the factors with the strongest correlation were the time studied, which had a positive correlation, and the number of classes failed, which had a negative correlation. 'Free Time Rating' and 'Family Relationship Rating' showed a positive correlation, while 'Health Status' showed a negative correlation, however, these plots showed that these correlations weren't strong, as the difference between the highest and lowest final grade was 2. The absences graph showed no clear signs of any trends, as grades randomly spiked in increase or decrease. Given that 'Time Studied' and 'Classes Failed' are both academic factors, this also shows that our hypothesis is incorrect, since these academic factors had a higher correlation with the final grades over the personal and mental health factors. Despite our findings, there were high uncertainty margins for all of our line plots.

As a result, we conclude that academic factors, specifically past class failures and weekly study time, have a larger impact on academic performance of students than personal and mental health factors.
